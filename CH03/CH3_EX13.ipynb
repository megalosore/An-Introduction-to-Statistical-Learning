```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ISLP
import statsmodels.api as sm
import seaborn as sns
from ISLP.models import (ModelSpec as MS,
                         summarize,
                         poly)
```

```python
rng = np.random.default_rng (1)
x = rng.normal(size =100)
eps = rng.normal(scale = 0.1 , size =100)
true_B0 = 1
true_B1 = 0.5
y = true_B0 + true_B1 * x + eps
```

```python
model = sm.OLS(y, list(zip([1]*100, x)));
results = model.fit();
sns.scatterplot(x=x, y=y)
sns.lineplot(x=x, y=results.params[0] + results.params[1]*x, color='r', linestyle='dashed')
plt.title("Scatterplot and Least Square Regression");
plt.xlabel("x");
plt.ylabel("y");
```

```python
f" The estimated intercept is : {results.params[0]}, The estimated coeff is {results.params[1]}, the error on the coefficients are : {results.params[0] - true_B0}, {true_B1 - results.params[1]}"
```

```python
results.summary()
```

```python
model = sm.OLS(y, list(zip([1]*100, x, x*x)));
results = model.fit();
sns.scatterplot(x=x, y=y)
sns.lineplot(x=x, y=results.params[0] + results.params[1]*x + results.params[2]*x*x, color='r', linestyle='dashed')
plt.title("Scatterplot and Least Square Regression");
plt.xlabel("x");
plt.ylabel("y");
```
