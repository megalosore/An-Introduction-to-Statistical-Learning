```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
y = -6
lamb = 4
x_range = np.linspace(-10,10, num=1000)
def ridge_loss(b, lamb, y):
    return((y - b)**2 + lamb * b**2)
def lasso_loss(b, lamb, y):
    return ((y - b)**2 + lamb * abs(b))
```

```python
def best_lasso_beta(y, lamb):
    res = 0
    if y > lamb/2:
        res = y - 0.5 * lamb
    elif y < -lamb/2:
        res = y + 0.5 * lamb
    else:
        res = 0
    return res
```

```python
ridge_points = [ridge_loss(b, lamb, y) for b in x_range]
ridge_best_b = y/(1 + lamb)
ridge_best_rss = ridge_loss(best_b, lamb, y)
```

```python
lasso_points = [lasso_loss(b, lamb, y) for b in x_range]
lasso_best_b = best_lasso_beta(y, lamb)
lasso_best_rss = lasso_loss(best_b, lamb, y)
```

```python
sns.scatterplot(y=ridge_points, x=x_range)
sns.scatterplot(y=lasso_points, x=x_range, color='orange', )
plt.axvline(lasso_best_b, color='orange', linestyle = '--')
plt.axvline(ridge_best_b, color='blue', linestyle = '--')
```
